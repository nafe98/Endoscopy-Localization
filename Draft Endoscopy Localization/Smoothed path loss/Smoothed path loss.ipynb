# Importing Libraries for Data Engineering

```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
sensors_data = pd.read_excel('Smoothed path loss of 48 sensors for 2443 positions.xlsx', header=None)
```

```python
sensors_data
```

```python
position_data = pd.read_excel('real x-y-z positions of 2443 locations.xlsx', header=None)
```

```python
position_data
```

# Data Engineering

### Sensors Data -- Labeling Columns 

```python
Sensors_Number_of_Columns = sensors_data.shape[1]
```

```python
Sensors_columns = ["sensor" + str(x) for x in range(1,Sensors_Number_of_Columns + 1)]
sensors_data.columns = (Sensors_columns)
```

```python
sensors_data
```

### Converting to Pandas Dataframe

```python
sensors_data = pd.DataFrame(sensors_data)
```

### Position Data -- Labeling Columns

```python
position_data.columns = ['Pos X', 'Pos Y', 'Pos Z']
```

```python
position_data
```

### Converting to Pandas Dataframe

```python
position_data = pd.DataFrame(position_data)
```

# Concatenating Dataframes

```python
raw_data = pd.concat([sensors_data, position_data], axis = 1)
raw_data
```
