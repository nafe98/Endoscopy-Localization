# Tuning Random Forest Regressor

# Importing Libraries

```python
import time
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error
from sklearn.metrics import explained_variance_score, max_error, mean_squared_log_error
```

# Initializing Kfold

```python
kfold = KFold(n_splits=20, shuffle=True, random_state=42)
```

# Getting Parameters 

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
print(rf.get_params())
```

# Adding Random Parameters

```python
param_grid = {
    'ccp_alpha': [0.0],
    'max_depth': [None, 5, 10, 15],
    'max_features': ['auto', 'sqrt', 'log2', 0.25, 0.5, 0.75, 1.0],
    'max_leaf_nodes': [None, 5, 10, 20],
    'max_samples': [None],
    'min_impurity_decrease': [0.0, 0.01, 0.1],
    'min_samples_leaf': [1, 2, 4, 8],
    'min_samples_split': [2, 5, 10, 20],
    'min_weight_fraction_leaf': [0.0, 0.1, 0.2],
    'n_estimators': [100, 200, 300]
}
```

# Define hyperparameters to search for each model

```python
models_params = [(RandomForestRegressor(),
                  
{
    'ccp_alpha': [0.0],
    'max_depth': [None, 5, 10, 15],
    'max_features': ['auto', 'sqrt', 'log2', 0.25, 0.5, 0.75, 1.0],
    'max_leaf_nodes': [None, 5, 10, 20],
    'max_samples': [None],
    'min_impurity_decrease': [0.0, 0.01, 0.1],
    'min_samples_leaf': [1, 2, 4, 8],
    'min_samples_split': [2, 5, 10, 20],
    'min_weight_fraction_leaf': [0.0, 0.1, 0.2],
    'n_estimators': [100, 200, 300]
    
}
                  
                 )]
```

```python
# Create empty dictionary to store the results
results_dict = {
    "Model": [],
    "Fold": [],
    "Mean Squared Error (MSE)": [],
    "Mean Absolute Error (MAE)": [],
    "Root Mean Squared Error (RMSE)": [],
    "Time taken": [],
    "Best Parameters": []
}
```

# Loop over models and perform hyperparameter tuning

```python
for model, params in models_params:
    
    #GridSearchCV
    
    
    start_time = time.time()
    grid_search = GridSearchCV(model, params, cv=kfold, scoring='neg_mean_squared_error', n_jobs=-1)
    grid_search.fit(X, y)
    end_time = time.time()
    results_dict["Model"].append(str(model).split('(')[0])
    results_dict["Fold"].append("Grid Search")
    results_dict["Mean Squared Error (MSE)"].append(-grid_search.best_score_)
    results_dict["Mean Absolute Error (MAE)"].append(mean_absolute_error(y, grid_search.predict(X)))
    results_dict["Root Mean Squared Error (RMSE)"].append(np.sqrt(-grid_search.best_score_))
    results_dict["Time taken"].append(end_time - start_time)
    results_dict["Best Parameters"].append(grid_search.best_params_)
    
    #RandomizedSearchCV
    
    
    start_time = time.time()
    random_search = RandomizedSearchCV(model, params, cv=kfold, scoring='neg_mean_squared_error', n_jobs=-1)
    random_search.fit(X, y)
    end_time = time.time()
    results_dict["Model"].append(str(model).split('(')[0])
    results_dict["Fold"].append("Randomized Search")
    results_dict["Mean Squared Error (MSE)"].append(-random_search.best_score_)
    results_dict["Mean Absolute Error (MAE)"].append(mean_absolute_error(y, random_search.predict(X)))
    results_dict["Root Mean Squared Error (RMSE)"].append(np.sqrt(-random_search.best_score_))
    results_dict["Time taken"].append(end_time - start_time)
    results_dict["Best Parameters"].append(random_search.best_params_)
```

# Display Results

```python
# set display options
pd.set_option("display.max_columns", None)  # show all columns
pd.set_option("display.width", None)  # auto-adjust column width
pd.set_option("display.max_colwidth", None)  # show full column contents

# Create a pandas dataframe from the results dictionary
results_df = pd.DataFrame(results_dict)

# Group the results by model and fold, and display the best parameters for each group
grouped_results = results_df.groupby(["Model", "Fold"]).apply(lambda x: x.loc[x["Mean Squared Error (MSE)"].idxmin()]).reset_index(drop=True)

# Create a new column with the Best Parameters in full sentence format
grouped_results["Full Parameters"] = grouped_results["Best Parameters"].apply(lambda x: ', '.join(['{}={}'.format(key, val) for key, val in x.items()]))

# Display the grouped results with the Full Parameters column
display(grouped_results[["Model", "Fold", "Full Parameters", "Mean Squared Error (MSE)", "Mean Absolute Error (MAE)", "Root Mean Squared Error (RMSE)", "Time taken"]])
```
