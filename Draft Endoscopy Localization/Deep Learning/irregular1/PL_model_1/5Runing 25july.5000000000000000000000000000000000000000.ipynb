# Importing Libraries for Data Engineering

```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
sensors_data = pd.read_excel('PL_model_1_Scattered_iReg1.xlsx', header=None)
```

```python
position_data = pd.read_excel('real_positions_iReg1.xlsx', header=None)
```

# Data Engineering

### Sensors Data -- Labeling Columns 

```python
Sensors_Number_of_Columns = sensors_data.shape[1]
```

```python
Sensors_columns = ["sensor" + str(x) for x in range(1,Sensors_Number_of_Columns + 1)]
sensors_data.columns = (Sensors_columns)
```

```python
sensors_data
```

### Converting to Pandas Dataframe

```python
sensors_data = pd.DataFrame(sensors_data)
```

### Position Data -- Labeling Columns

```python
position_data.columns = ['Pos X', 'Pos Y', 'Pos Z']
```

```python
position_data
```

### Converting to Pandas Dataframe

```python
position_data = pd.DataFrame(position_data)
```

# Split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sensors_data, position_data, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

# Importing Deep Learning Libraries

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam,RMSprop

from keras.callbacks import EarlyStopping
from tensorflow.keras.layers import BatchNormalization
```

# Deep Learning Architecture

Sure, to adapt the given architecture for a regression problem, you can add appropriate activation functions and batch normalization layers. For regression problems, you typically don't use activation functions in the output layer since you want the model to directly predict continuous values. However, you can use activation functions in the hidden layers to introduce non-linearity and improve the model's ability to learn complex patterns.

Also, batch normalization can help in improving the training process by normalizing the input to each layer, which aids in faster convergence and reduces the dependence on the initialization of the weights. Here's an updated version of the architecture with activation functions and batch normalization:

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, BatchNormalization, Dense, Activation
from keras.optimizers import RMSprop
from keras.models import Sequential
from keras.layers import LSTM, BatchNormalization, Activation, Dense, Dropout
```

```python
model = Sequential()
model.add(LSTM(512, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.3))  

model.add(LSTM(256, return_sequences=True))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.3)) 

model.add(LSTM(128))
model.add(Activation('relu'))


model.add(Dense(64))
model.add(Dense(3))



adam = Adam(lr=0.0001)
model.compile(loss='mean_squared_error', optimizer=adam) 


model.fit(X_train, y_train, epochs=200, batch_size=5, validation_data=(X_test, y_test)) 

mse = model.evaluate(X_test, y_test) 
print("MSE:", mse)
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Predict on test data
y_pred = model.predict(X_test)
```

```python
# Calculate evaluation metrics
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("="*100)
```
