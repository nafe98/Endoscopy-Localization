# Importing Libraries for Data Engineering

```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
sensors_data = pd.read_excel('Smoothed path loss of 48 sensors for 2443 positions.xlsx', header=None)
```

```python
position_data = pd.read_excel('real x-y-z positions of 2443 locations.xlsx', header=None)
```

```python
position_data
```

# Data Engineering

### Sensors Data -- Labeling Columns 

```python
Sensors_Number_of_Columns = sensors_data.shape[1]
```

```python
Sensors_columns = ["sensor" + str(x) for x in range(1,Sensors_Number_of_Columns + 1)]
sensors_data.columns = (Sensors_columns)
```

```python
sensors_data
```

### Converting to Pandas Dataframe

```python
sensors_data = pd.DataFrame(sensors_data)
```

### Position Data -- Labeling Columns

```python
position_data.columns = ['Pos X', 'Pos Y', 'Pos Z']
```

```python
position_data
```

### Converting to Pandas Dataframe

```python
position_data = pd.DataFrame(position_data)
```

# Concatenating Dataframes

```python
raw_data = pd.concat([sensors_data, position_data], axis = 1)
raw_data
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# load the data into a pandas DataFrame
data = raw_data

# extract the features (X) and target variable (y)
X = data.iloc[:, :-3]
y = data.iloc[:, 0]

# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# create a list of the models to be used
models = [
    LinearRegression(),
    DecisionTreeRegressor(),
    SVR(),
    KNeighborsRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor()
]

# fit the models to the training data and make predictions on test data
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    # calculate the evaluation metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    # print the results for each model
    print(f"Model: {model.__class__.__name__}")
    print(f"Mean Squared Error: {mse:.4f}")
    print(f"Root Mean Squared Error: {rmse:.4f}")
    print(f"Mean Absolute Error: {mae:.4f}")
    print(f"R^2 Score: {r2:.4f}")
    print("\n")
```
