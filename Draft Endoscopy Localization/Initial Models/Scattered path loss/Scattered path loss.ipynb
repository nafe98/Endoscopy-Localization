```python
regressor = Sequential()
# First LSTM layer with Dropout regularization
regressor.add(LSTM(units=450, return_sequences=True, input_shape=(X_train.shape[1],1), dropout=0.2))

# Second LSTM layer
regressor.add(LSTM(units=100, return_sequences=False, dropout=0.2))

# The output layer
regressor.add(Dense(units=3))

# Compiling the RNN
regressor.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
# Fitting to the training set
regressor.fit(X_train,y_train,epochs=50,batch_size=32)
```

```python
# Define the model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(X_train.shape[1],1)))
model.add(Dense(3))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=2)

# Evaluate the model on the test data
test_error = model.evaluate(X_test, y_test, verbose=0)
print("Test MSE: ", test_error)
```

```python
import matplotlib.pyplot as plt

# Predict the values for the training and testing sets
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Plot the training and testing predictions against the true values
plt.plot(y_train, label='True Values (Training)')
plt.plot(train_predict, label='Predicted Values (Training)')
plt.legend()
plt.show()

plt.plot(y_test, label='True Values (Testing)')
plt.plot(test_predict, label='Predicted Values (Testing)')
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

def create_lstm_model(X_values, y_values, epochs=50, batch_size=32, verbose=2):
    """
    Creates and trains an LSTM model for a regression problem.
    
    Arguments:
    X_values -- 2D numpy array of input values
    y_values -- 2D numpy array of target values
    epochs -- number of training epochs (default: 50)
    batch_size -- batch size for training (default: 32)
    verbose -- verbosity mode during training (default: 2)
    
    Returns:
    model -- trained LSTM model
    """
    # Split the data into training and testing sets
    X_train = X_values[:2000, :]
    X_test = X_values[2000:, :]
    y_train = y_values[:2000, :]
    y_test = y_values[2000:, :]

    # Reshape the data for the LSTM model
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Define the model
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dense(3))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # Train the model
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=verbose)

    # Evaluate the model on the test data
    test_error = model.evaluate(X_test, y_test, verbose=0)
    print("Test MSE: ", test_error)
    
    return model
```
