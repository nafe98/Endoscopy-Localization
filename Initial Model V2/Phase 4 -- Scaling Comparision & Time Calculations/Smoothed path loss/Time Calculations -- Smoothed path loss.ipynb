# Importing Libraries for Data Engineering

```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
sensors_data = pd.read_excel('Smoothed path loss of 48 sensors for 2443 positions.xlsx', header=None)
```

```python
sensors_data
```

```python
position_data = pd.read_excel('real x-y-z positions of 2443 locations.xlsx', header=None)
```

```python
position_data
```

# Data Engineering

### Sensors Data -- Labeling Columns 

```python
Sensors_Number_of_Columns = sensors_data.shape[1]
```

```python
Sensors_columns = ["sensor" + str(x) for x in range(1,Sensors_Number_of_Columns + 1)]
sensors_data.columns = (Sensors_columns)
```

```python
sensors_data
```

### Converting to Pandas Dataframe

```python
sensors_data = pd.DataFrame(sensors_data)
```

### Position Data -- Labeling Columns

```python
position_data.columns = ['Pos X', 'Pos Y', 'Pos Z']
```

```python
position_data
```

### Converting to Pandas Dataframe

```python
position_data = pd.DataFrame(position_data)
```

### Concatenating Dataframes

```python
raw_data = pd.concat([sensors_data, position_data], axis = 1)
raw_data
```

```python
# get the signal values
signal = df.iloc[:, :48].values

# create a figure and axis objects
fig, axs = plt.subplots(8, 6, figsize=(15, 15), sharex=True, sharey=True)

# plot a line plot for each signal
for i in range(48):
    row = i // 6
    col = i % 6
    axs[row, col].plot(signal[:, i], label=f"Signal {i+1}", linewidth=0.7)
    axs[row, col].set_title(f"Signal {i+1}", fontsize=10, fontweight='bold')
    axs[row, col].tick_params(axis='both', which='major', labelsize=8)
    axs[row, col].legend().remove()

# remove empty subplots
for i in range(48, 48):
    fig.delaxes(axs[i//6, i%6])

# add labels to only the bottom and leftmost plots
for i in range(0, 6):
    axs[7, i].set_xlabel("Sample", fontsize=10)
for i in range(8):
    axs[i, 0].set_ylabel("Signal Value", fontsize=10)

# add a common legend
fig.legend(loc='center', bbox_to_anchor=(0.5, 0.05), ncol=6, fontsize=10)

# add grid to all plots
for ax in axs.flat:
    ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# adjust spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# show plot
plt.show()
```

# Split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sensors_data, position_data, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

# Vanilla Machine Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error
from sklearn.metrics import explained_variance_score, max_error, mean_squared_log_error
```

```python
# Initialize the models
models = [LinearRegression(), DecisionTreeRegressor(), RandomForestRegressor(), KNeighborsRegressor(),  XGBRegressor()]
```

```python
# Train and evaluate each model
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(type(model).__name__)
    print("Accuracy:", model.score(X_test, y_test))
    print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
    print("R2 Score:", r2_score(y_test, y_pred))
    print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
    print("Median Absolute Error:", median_absolute_error(y_test, y_pred))
    print("Explained Variance Score:", explained_variance_score(y_test, y_pred))
    print("="*100)
```

```python
# Evaluate each model
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(type(model).__name__, " accuracy: ", model.score(X_test, y_test))
    fig, ax = plt.subplots()
    ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
    ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
    ax.set_xlabel('Actual')
    ax.set_ylabel('Predicted')
    plt.show()
```

# Time Calculations

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error, explained_variance_score
import time

# Train and evaluate each model
for model in models:
    start_time = time.time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    end_time = time.time()
    
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    
    print(type(model).__name__)
    print("Accuracy:", model.score(X_test, y_test))
    print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
    print("Root Mean Squared Error:", rmse)
    print("R2 Score:", r2_score(y_test, y_pred))
    print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
    print("Median Absolute Error:", median_absolute_error(y_test, y_pred))
    print("Explained Variance Score:", explained_variance_score(y_test, y_pred))
    print("Time taken:", end_time - start_time)
    print("="*100)
```
